```python
#4 ESTRUTURAS DE LAÇOS E CONDICIONAIS
# Escolha QUANTITYORDERED
import pandas as pd

df = pd.read_csv("sales_data_sample (1).csv", encoding='latin-1')


#ESTRUTURA CONDICIONAL


coluna_numerica = 'QUANTITYORDERED'
primeiro_valor = df[coluna_numerica].iloc[0]

print("Analisando o primeiro valor da coluna QUANTITYORDERED:", primeiro_valor)

if primeiro_valor > 40:
    print("→ O valor é ALTO")
elif primeiro_valor > 20:
    print("→ O valor é INTERMEDIÁRIO")
else:
    print("→ O valor é BAIXO")


#LAÇO FOR: somando os 5 primeiros valores


soma = 0
lista = list(df[coluna_numerica].iloc[:5])

for n in lista:
    soma += n

print("Soma dos 5 primeiros valores:", soma)


#LAÇO WHILE

limite = 35
i = 0

while i < len(lista) and lista[i] <= limite:
    i += 1

if i < len(lista):
    print(f"Primeiro valor > {limite} encontrado:", lista[i])
else:
    print(f"Nenhum valor maior que {limite} foi encontrado nos primeiros 5.")
```


Insight

O primeiro pedido teve quantidade 30, o que foi considerado INTERMEDIÁRIO.

A soma das 5 primeiras quantidades é 199 (30 + 34 + 41 + 45 + 49), sugerindo um volume inicial alto.

Resumo: O volume inicial é alto devido à soma das 5 primeiras quantidades ser 199.

Recomendação: Verificar os pedidos subsequentes para confirmar a tendência de volume.



```python
#5 USANDO OPERADORES ARITMÉTICOS E MANIPULAÇÃO DA DADOS
import pandas as pd

df = pd.read_csv("sales_data_sample (1).csv", encoding='latin-1')


#Cálculo da diferença entre preço cheio e preço pago

if 'MSRP' in df.columns and 'PRICEEACH' in df.columns:
    df['Dif_Precos'] = df['MSRP'] - df['PRICEEACH']
    print(" Diferença entre preço cheio e preço pago:")
    print(df[['MSRP', 'PRICEEACH', 'Dif_Precos']].head())


# Criando nova coluna: preço com 10% de desconto aplicado

if 'PRICEEACH' in df.columns:
    df['PRECO_DESC_10%'] = df['PRICEEACH'] * 0.9
    print("\n Preço com desconto de 10%:")
    print(df[['PRICEEACH', 'PRECO_DESC_10%']].head())


# Criando nova coluna: quantidade * preço com desconto = receita estimada promocional

if 'QUANTITYORDERED' in df.columns and 'PRECO_DESC_10%' in df.columns:
    df['SALES_PROMO'] = df['QUANTITYORDERED'] * df['PRECO_DESC_10%']
    print("\n Receita estimada com desconto aplicado:")
    print(df[['QUANTITYORDERED', 'PRECO_DESC_10%', 'SALES_PROMO']].head())
```

 Insight:

A média da diferença entre o **MSRP** (preço sugerido) e o **preço real de venda** foi de aproximadamente **\$3.31**.  
Isso mostra que, em média, os produtos foram vendidos **abaixo do valor sugerido**, com variações tanto positivas quanto negativas.

Foram identificados **96 produtos** com desconto, indicando que **promoções ou negociações** são práticas frequentes na política de preços.

A nova coluna **`SALES_PROMO`** estima o faturamento **caso todos os pedidos tivessem 10% de desconto aplicado** — o que ajuda a simular cenários de campanhas promocionais e avaliar o impacto no faturamento.

 Essa análise sugere que a empresa já oferece descontos pontuais, e que a adoção de uma política promocional uniforme pode ser analisada com base no impacto estimado pela coluna `SALES_PROMO`.

O agrupamento por linha de produto revelou que a categoria **'Classic Cars'** é disparada a que mais vende em volume, seguida por **'Motorcycles'** e **'Planes'**. Isso sugere que esses produtos têm maior giro e podem receber atenção especial em estratégias comerciais.

Além disso, a contagem de frequência mostrou que há uma distribuição equilibrada entre as categorias, mas algumas aparecem bem menos, como **'Trucks and Buses'**, indicando nichos mais específicos.

**Conclusão**: As categorias com maior volume também são as mais frequentes, o que reforça a importância de manter estoque regular para “Classic Cars” e similares. Já categorias menos populares podem ser exploradas em ações de marketing para reduzir estoque parado.


```python
#8 VISUALIZAÇÃO DE DADOS
import pandas as pd
import matplotlib.pyplot as plt

# Carrega os dados
df = pd.read_csv("sales_data_sample (1).csv", encoding="latin-1")

# Gráfico de Linha: tendência de vendas por mês (construção de data)
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
df['AnoMes'] = df['ORDERDATE'].dt.to_period('M')
vendas_mensais = df.groupby('AnoMes')['SALES'].sum()

vendas_mensais.plot(kind='line', marker='o', figsize=(10, 5))
plt.title('Evolução Mensal das Vendas')
plt.xlabel('Ano-Mês')
plt.ylabel('Total de Vendas ($)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Gráfico de Barras: vendas por linha de produto
df.groupby('PRODUCTLINE')['SALES'].sum().sort_values().plot(kind='barh', color='skyblue')
plt.title('Total de Vendas por Categoria de Produto')
plt.xlabel('Total de Vendas ($)')
plt.ylabel('Categoria')
plt.tight_layout()
plt.show()

# Gráfico de Dispersão: preço x quantidade
plt.figure(figsize=(8, 5))
plt.scatter(df['PRICEEACH'], df['QUANTITYORDERED'], alpha=0.5, color='green')
plt.title('Dispersão: Preço Unitário x Quantidade Vendida')
plt.xlabel('Preço Unitário ($)')
plt.ylabel('Quantidade Vendida')
plt.grid(True)
plt.tight_layout()
plt.show()

```

**Gráfico de Linha: Evolução Mensal das Vendas**

O gráfico de linha mostra uma **tendência de sazonalidade** nas vendas. Há picos visíveis em determinados meses, sugerindo influência de datas comemorativas ou períodos promocionais. Isso pode embasar a programação de campanhas de marketing e estoque.

---

**Gráfico de Barras: Vendas por Categoria de Produto**

É evidente que a categoria **"Classic Cars"** lidera em volume de vendas, seguida por **"Motorcycles"** e **"Planes"**. Isso reforça o padrão identificado anteriormente e destaca quais linhas devem receber mais atenção em termos de reposição e divulgação.

---

**Gráfico de Dispersão: Preço x Quantidade Vendida**

O gráfico mostra uma **tendência inversa sutil**: produtos com **preço mais baixo tendem a ser vendidos em maiores quantidades**, enquanto itens mais caros têm vendas unitárias mais baixas. Isso confirma um padrão típico de elasticidade-preço nas vendas.

 **Conclusão Geral**: As visualizações reforçam os achados anteriores sobre desempenho por categoria e mostram como preço e quantidade interagem. Isso pode ser utilizado para montar estratégias de precificação e promoção.
